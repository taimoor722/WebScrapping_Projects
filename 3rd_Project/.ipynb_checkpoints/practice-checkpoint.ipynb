```python
import requests
import re
page=requests.get('https://partscatalog.deere.com/jdrc/search/type/model/term/70')
from bs4 import BeautifulSoup as bsup
element1_html=bsup(google.page_source,'lxml')
element1_html
# element1_html_item=element1_html.find_all('div',{'class':'row partresultsection widthLimit ng-star-inserted'})
element1_html_item=element1_html.find_all(lambda tag:tag.name=='a'and tag.get('class')==["ng-star-inserted"])
element1_html_item=element1_html.select('a[class=ng-star-inserted]')
element1_html_item=set(element1_html_item)
ls=[]
for i in element1_html_item:
    ls.append(re.findall(r'>([^.]*)<br',str(i) )[0].lstrip())
    print(re.findall(r'>([^.]*)<br',str(i) ) )
ls
# element1_html_item
```

```python
not_available=[]
wait.until(presence_of_element_located(   (By.XPATH,'//*[@id="justacontainer"]/div[3]/p')  )    )

for i in range(1,len(ls)+1):
    google.get('https://partscatalog.deere.com/jdrc/search/type/model/term/70')
    wait.until(presence_of_element_located(   (By.XPATH,'//*[@id="justacontainer"]/div[3]/p')  )    )

    element1=google.find_element_by_xpath(f"//*[@id='applicationContainer']/div/div[2]/div[1]/div[2]/app-search/div/div[2]/app-search-results/div/div[3]/a[{i}]")
    print(element1.text)
    element1.click()
    
    try:
        print('inside try')
        wait.until(presence_of_element_located(   (By.XPATH,'//*[@id="applicationContainer"]/div/div[2]/div[1]/div[2]/app-navigation/div/div[2]/app-navigation-results/div/div[1]/div[3]/a')  )    )
        print('waiting end')
        element11=google.find_element_by_xpath('//*[@id="applicationContainer"]/div/div[2]/div[1]/div[2]/app-navigation/div/div[2]/app-navigation-results/div/div[1]/div[3]/a')
        element11.click()
        print('try run')
        
        wait.until(presence_of_element_located(   (By.XPATH,'//*[@id="applicationContainer"]/div/div[2]/div[1]/div[2]/app-sidebyside/div/div/div[2]/div[4]')  )    )
        if 'pgId' in google.current_url: 
            items=google.find_element_by_xpath('//*[@id="applicationContainer"]/div/div[2]/div[1]/div[2]/app-sidebyside/div/div/div[2]/div[4]').text
            items
            data_ls=re.split('\nSelect Dealer to See Price and Availability\nQuantity\nAdd to Cart\n',items)
            for i in data_ls:
                if re.split("\d+ - |Part Number|Qty Req|Serial No.|Remarks",i,flags=re.IGNORECASE)[-1] in ['NO LONGER AVAILABLE','Nil']:
                    not_available.append(i)
            print('not_available',not_available)
        else:
            while 'pgid' not in google.current_url:
                google.find_element_by_xpath('//*[@id="pictorialCalloutList"]/a[1]').click()
                
                if 'pgId' in google.current_url: 
                    items=google.find_element_by_xpath('//*[@id="applicationContainer"]/div/div[2]/div[1]/div[2]/app-sidebyside/div/div/div[2]/div[4]').text
                    items
                    data_ls=re.split('\nSelect Dealer to See Price and Availability\nQuantity\nAdd to Cart\n',items)
                    for i in data_ls:
                        if re.split("\d+ - |Part Number|Qty Req|Serial No.|Remarks",i,flags=re.IGNORECASE)[-1] in ['NO LONGER AVAILABLE','Nil']:
                            not_available.append(i)
                    print('not_available',not_available)
            
    except Exception as e:
        print('inside  except',e)
#         wait.until(presence_of_element_located(   (By.XPATH,'//*[@id="applicationContainer"]/div/div[2]/div[1]/div[2]/app-navigation/div/div[2]/app-navigation-results/div/div[1]/div[1]/a[1]')  )    )

        element11=google.find_element_by_xpath('//*[@id="applicationContainer"]/div/div[2]/div[1]/div[2]/app-navigation/div/div[2]/app-navigation-results/div/div[1]/div[1]/a[1]')
        element11.click()
        print('except run')
        
        items=google.find_element_by_xpath('//*[@id="applicationContainer"]/div/div[2]/div[1]/div[2]/app-sidebyside/div/div/div[2]/div[4]').text
        items
        data_ls=re.split('\nSelect Dealer to See Price and Availability\nQuantity\nAdd to Cart\n',items)
        for i in data_ls:
            if re.split("\d+ - |Part Number|Qty Req|Serial No.|Remarks",i,flags=re.IGNORECASE)[-1] in ['NO LONGER AVAILABLE','Nil']:
                not_available.append(i)
        print('not_available',not_available)
        break

        
```
