```python
import requests
import re
page=requests.get('https://partscatalog.deere.com/jdrc/search/type/model/term/70')
from bs4 import BeautifulSoup as bsup
element1_html=bsup(google.page_source,'lxml')
element1_html
element1_html_item=element1_html.find_all('div',{'class':'row partresultsection widthLimit ng-star-inserted'})
# element1_html_item=element1_html.find_all(lambda tag:tag.name=='a'and tag.get('class')==["ng-star-inserted"])
element1_html_item=element1_html.select('a[class=ng-star-inserted]')
# element1_html_item=set(element1_html_item)
ls=[]
c=0
for i in element1_html_item:
    c+=1
    ls.append(re.findall(r'>([^.]*)<br',str(i) )[0].lstrip())
#     print(re.findall(r'>([^.]*)<br',str(i) ) )
print(c,len(ls),sep='-')
for index,elements in enumerate(ls,start=1):
    print(index,'-->',elements)
    
# element1_html_item
```

```python
run=True
while run:
    starting=int(input('select from above starting point using given no 1-22 :'))
    ending = int(input('select from above ending point using given no 1-22 :'))
    
    if starting <= ending:
        if starting in range(1,len(ls)+1) and ending in range(1,len(ls)+1):
            run=False
        else:
            print(f'both should be from {1}-{len(ls)}')
            run=True
    else:
        print('starting should be less than ending')
        run=True
```
